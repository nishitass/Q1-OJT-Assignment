Data Visualization techniques

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=iris_data, x="Sepal.Length", bins=10, kde=True)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=iris_data, x=iris_data.index, y="Sepal.Length")
plt.title("Line Chart of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
sns.countplot(data=iris_data, x="Species")
plt.title("Bar Chart of Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_data)
plt.title("Box Plot of Iris Dataset")
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_data, x="Sepal.Length", y="Petal.Length", hue="Species")
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
# Exclude the 'species' column before calculating the correlation matrix
numerical_columns = iris_data.drop(columns='Species')
correlation_matrix = numerical_columns.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
iris_data['Species'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart of Species Distribution")
plt.ylabel("")
plt.show()
```

```python
# Area chart (not meaningful for this dataset, using a placeholder)
plt.figure(figsize=(10, 6))
plt.fill_between(range(len(iris_data)), iris_data["Sepal.Length"], alpha=0.5)
plt.title("Area Chart of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris_data)
plt.title("Violin Plot of Iris Dataset")
plt.show()
```

```python
# Bubble chart (not meaningful for this dataset, using a placeholder)
plt.figure(figsize=(10, 6))
plt.scatter(iris_data.index, iris_data["Sepal.Length"], s=iris_data["Petal.Length"]*10, alpha=0.5)
plt.title("Bubble Chart of Sepal Length vs Petal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

Statistical measures

```python
print("\nStatistical measures for each numerical column in the Iris dataset:")
statistical_measures = iris_data.describe()
print(statistical_measures)
```

```python
# Calculate mode for each numerical column
modes = iris_data.mode()
modes = modes.dropna(axis=1, how='all')  # Drop columns with all NaN values
modes = modes.drop(0)  # Drop the row with index 0 (contains column names)
modes = modes.rename(index={1: 'mode'})  # Rename index to 'mode'
print("\nMode for each numerical column in the Iris dataset:")
print(modes)
```

```python
# Calculate range for each numerical column
numerical_columns = iris_data.select_dtypes(include=['float64', 'int64'])
# Select only numerical columns
ranges = numerical_columns.max() - numerical_columns.min()
print("\nRange for each numerical column in the Iris dataset:")
print(ranges)
```

```python
# Calculate variance for each numerical column
numerical_columns = iris_data.select_dtypes(include=['float64', 'int64'])
# Select only numerical columns
variances = numerical_columns.var()
print("\nVariance for each numerical column in the Iris dataset:")
print(variances)
```

```python
# Calculate quartiles for each numerical column
numerical_columns = iris_data.select_dtypes(include=['float64', 'int64'])
# Select only numerical columns
quartiles = numerical_columns.quantile([0.25, 0.5, 0.75])
print("\nQuartiles for each numerical column in the Iris dataset:")
print(quartiles)
```

```python
# Calculate Interquartile Range (IQR) for each numerical column

IQR = quartiles.loc[0.75] - quartiles.loc[0.25]
print("\nInterquartile Range (IQR) for each numerical column in the Iris dataset:")
print(IQR)
```

```python
# Calculate skewness for each numerical column

numerical_columns = iris_data.select_dtypes(include=['float64', 'int64'])
# Select only numerical columns
skewness = numerical_columns.skew()
print("\nSkewness for each numerical column in the Iris dataset:")
print(skewness)
```

```python
# Calculate kurtosis for each numerical column

numerical_columns = iris_data.select_dtypes(include=['float64', 'int64'])
# Select only numerical columns
kurtosis = numerical_columns.kurtosis()
print("\nKurtosis for each numerical column in the Iris dataset:")
print(kurtosis)
```
